```python
import random
def random_number():
    return random.randint(290500,302500)
def game(num,chances):
    has_chances = True
    while(has_chances):
        
        if(chances == 0):
            print(f"You have run out of guesses.The correct number is {num}. YOU LOST")
            break
        else:
            print(f"You have {chances} chances remaining to guess the number")
            guessed_number = int(input("Make a guess:"))
        
        if(guessed_number<number):
            chances -= 1
            if(chances != 0):
                print("Too low")
                print("Guess again")
        elif(guessed_number>number):
            chances -= 1
            if(chances!=0):
                print("Too high")
                print("Guess again")
        else:
            print(f"Your guess is correct. YOU WIN!")
            has_chances = False
print("Welcome to the NUMBER GUESSING GAME!")
print("I'm thinking of a number between 290500 to 302500")
level = input("Choose a difficulty level. Type 'easy' or 'medium' or 'hard'")
number = random_number()
if(level == "easy"):
    game(number,20)
elif(level == "medium"):
    game(number,12)
elif(level == "hard"):
    game(number,7)
else:
    print("Invalid input")
    exit
```
